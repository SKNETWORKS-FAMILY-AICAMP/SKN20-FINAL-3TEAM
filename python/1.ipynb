```python
# 법령 데이터 로드 (이미 로드되어 있다면 data2 사용)
if 'data2' not in globals():
    data2 = pd.read_csv('csv/법령_전처리2(특수문자 제거).csv')

# 법령 데이터의 용도지역지구명 고유 값
law_unique = set(data2['용도지역지구명'].unique())
ordinance_unique = set(data['용도지역지구명'].unique())

print(f"법령 고유 값 갯수: {len(law_unique)}")
print(f"조례 고유 값 갯수: {len(ordinance_unique)}")

# 공통 값
common = law_unique & ordinance_unique
print(f"공통 값 갯수: {len(common)}")
print("공통 값 샘플:", list(common)[:10])

# 법령에만 있는 값
law_only = law_unique - ordinance_unique
print(f"법령에만 있는 값 갯수: {len(law_only)}")
print("법령에만 있는 값 샘플:", list(law_only)[:10])

# 조례에만 있는 값
ord_only = ordinance_unique - law_unique
print(f"조례에만 있는 값 갯수: {len(ord_only)}")
print("조례에만 있는 값 샘플:", list(ord_only)[:10])

# 비슷한 값 찾기 (간단한 방법: 공통 단어로 그룹화)
from collections import defaultdict

def group_by_common_words(values):
    groups = defaultdict(list)
    for val in values:
        words = str(val).split()
        key = tuple(sorted(words))  # 단어 정렬해서 키로
        groups[key].append(val)
    return groups

law_groups = group_by_common_words(law_only)
ord_groups = group_by_common_words(ord_only)

print("\n법령에만 있는 값 그룹 샘플:")
for key, vals in list(law_groups.items())[:5]:
    if len(vals) > 1:
        print(f"그룹 {key}: {vals}")

print("\n조례에만 있는 값 그룹 샘플:")
for key, vals in list(ord_groups.items())[:5]:
    if len(vals) > 1:
        print(f"그룹 {key}: {vals}")
```

```python
# 비슷한 값 찾기 (유사도 기반)
from difflib import SequenceMatcher

def find_similar_pairs(list1, list2, threshold=0.8):
    similar = []
    for a in list1:
        for b in list2:
            sim = SequenceMatcher(None, str(a), str(b)).ratio()
            if sim >= threshold and a != b:
                similar.append((a, b, sim))
    return sorted(similar, key=lambda x: x[2], reverse=True)

# 법령에만 있는 값과 조례에만 있는 값 사이의 비슷한 쌍 찾기
similar_pairs = find_similar_pairs(list(law_only)[:50], list(ord_only)[:50], 0.7)  # 샘플로 50개씩

print(f"비슷한 쌍 갯수: {len(similar_pairs)}")
print("비슷한 쌍 샘플:")
for pair in similar_pairs[:10]:
    print(f"{pair[0]} <-> {pair[1]} (유사도: {pair[2]:.2f})")

# 매핑 딕셔너리 예시 (수동으로 추가 가능)
mapping_dict = {
    '준주거지구': '준주거지역',
    '상업지구': '일반상업지역',
    # 더 추가...
}

# 적용 예시
def apply_mapping(value, mapping):
    return mapping.get(str(value), str(value))

# 테스트
test_values = ['준주거지구', '상업지구', '일반상업지역']
mapped = [apply_mapping(v, mapping_dict) for v in test_values]
print(f"매핑 테스트: {test_values} -> {mapped}")
```

```python
# 자동 매핑 시도: '지구' -> '지역'으로 바꾸기
def standardize_name(name):
    name = str(name).strip()
    # '지구' -> '지역'
    name = name.replace('지구', '지역')
    # 다른 표준화 규칙 추가 가능
    name = name.replace('주거지구', '주거지역')  # 이미 위에서 했지만
    return name

# 법령 데이터 표준화
data2['용도지역지구명_표준'] = data2['용도지역지구명'].apply(standardize_name)
# 조례 데이터 표준화 (이미 data에 적용)
data['용도지역지구명_표준'] = data['용도지역지구명'].apply(standardize_name)

# 표준화 후 고유 값 비교
law_std_unique = set(data2['용도지역지구명_표준'].unique())
ord_std_unique = set(data['용도지역지구명_표준'].unique())

common_std = law_std_unique & ord_std_unique
print(f"표준화 후 공통 값 갯수: {len(common_std)}")
print("표준화 후 공통 값 샘플:", list(common_std)[:10])

law_std_only = law_std_unique - ord_std_unique
ord_std_only = ord_std_unique - law_std_unique
print(f"표준화 후 법령에만: {len(law_std_only)}, 조례에만: {len(ord_std_only)}")

# 이제 두 데이터의 용도지역지구명_표준을 합쳐서 value_counts
combined = pd.concat([data2[['용도지역지구명_표준']], data[['용도지역지구명_표준']]], ignore_index=True)
print("합친 데이터의 표준화된 용도지역지구명 value_counts 상위 20:")
print(combined['용도지역지구명_표준'].value_counts().head(20))
```

```python
# 재사용 가능한 코드: 조례 CSV 리스트와 법령 CSV를 입력받아 자동으로 처리
import pandas as pd
import os
from difflib import SequenceMatcher

def standardize_name(name):
    """용도지역지구명 표준화 함수"""
    name = str(name).strip()
    name = name.replace('지구', '지역')
    # 추가 규칙 가능
    return name

def combine_ordnances(csv_list):
    """조례 CSV 리스트를 하나로 합침"""
    dfs = []
    for csv in csv_list:
        if os.path.exists(csv):
            df = pd.read_csv(csv)
            dfs.append(df)
        else:
            print(f"파일 없음: {csv}")
    if dfs:
        combined = pd.concat(dfs, ignore_index=True)
        # 향토유적보호구역 통합
        combined['용도지역지구명'] = combined['용도지역지구명'].apply(
            lambda x: '향토유적보호구역' if str(x).endswith('향토유적보호구역') else x
        )
        return combined
    return pd.DataFrame()

def analyze_and_merge(ordinance_csv_list, law_csv):
    """메인 처리 함수"""
    # 1. 조례 데이터 합치기
    ord_data = combine_ordnances(ordinance_csv_list)
    if ord_data.empty:
        return "조례 데이터 없음"

    # 2. 법령 데이터 로드
    if not os.path.exists(law_csv):
        return f"법령 파일 없음: {law_csv}"
    law_data = pd.read_csv(law_csv)

    # 3. 표준화
    ord_data['용도지역지구명_표준'] = ord_data['용도지역지구명'].apply(standardize_name)
    law_data['용도지역지구명_표준'] = law_data['용도지역지구명'].apply(standardize_name)

    # 4. 고유 값 분석
    ord_unique = set(ord_data['용도지역지구명_표준'].unique())
    law_unique = set(law_data['용도지역지구명_표준'].unique())
    common = ord_unique & law_unique
    ord_only = ord_unique - law_unique
    law_only = law_unique - ord_unique

    # 5. 전체 합치기
    merged = pd.concat([law_data, ord_data], ignore_index=True, sort=False)

    # 6. 결과 저장
    merged.to_csv('final_merged_all_columns.csv', index=False)
    with open('ordinance_only.txt', 'w', encoding='utf-8') as f:
        f.write('\n'.join(sorted(list(ord_only))))
    with open('law_only.txt', 'w', encoding='utf-8') as f:
        f.write('\n'.join(sorted(list(law_only))))
    with open('common_values.txt', 'w', encoding='utf-8') as f:
        f.write('\n'.join(sorted(list(common))))

    return {
        '조례 행 수': len(ord_data),
        '법령 행 수': len(law_data),
        '합친 행 수': len(merged),
        '공통 값': len(common),
        '조례만': len(ord_only),
        '법령만': len(law_only),
        '저장 파일': ['final_merged_all_columns.csv', 'ordinance_only.txt', 'law_only.txt', 'common_values.txt']
    }

# 사용 예시 (파일 경로를 변경해서 사용)
# result = analyze_and_merge(['csv/조례1.csv', 'csv/조례2.csv'], 'csv/법령_전처리2(특수문자 제거).csv')
# print(result)
```

```python
# 파일 불러오기 함수 (CSV 또는 Excel 지원)
def load_file(file_path):
    if file_path.endswith(('.xlsx', '.xls')):
        return pd.read_excel(file_path)
    elif file_path.endswith('.csv'):
        return pd.read_csv(file_path)
    else:
        raise ValueError(f"지원하지 않는 파일 형식: {file_path}")

# combine_ordnances 함수 수정
def combine_ordnances(csv_list):
    """조례 CSV/Excel 리스트를 하나로 합침"""
    dfs = []
    for file in csv_list:
        if os.path.exists(file):
            df = load_file(file)
            dfs.append(df)
        else:
            print(f"파일 없음: {file}")
    if dfs:
        combined = pd.concat(dfs, ignore_index=True)
        # 향토유적보호구역 통합
        combined['용도지역지구명'] = combined['용도지역지구명'].apply(
            lambda x: '향토유적보호구역' if str(x).endswith('향토유적보호구역') else x
        )
        return combined
    return pd.DataFrame()

# analyze_and_merge 함수에서 법령 로드 수정
def analyze_and_merge(ordinance_file_list, law_file):
    """메인 처리 함수 (CSV/Excel 지원)"""
    # 1. 조례 데이터 합치기
    ord_data = combine_ordnances(ordinance_file_list)
    if ord_data.empty:
        return "조례 데이터 없음"

    # 2. 법령 데이터 로드
    if not os.path.exists(law_file):
        return f"법령 파일 없음: {law_file}"
    law_data = load_file(law_file)

    # 나머지 코드는 동일...
    # (이전 코드 그대로)

    return {
        '조례 행 수': len(ord_data),
        '법령 행 수': len(law_data),
        '합친 행 수': len(merged),
        '공통 값': len(common),
        '조례만': len(ord_only),
        '법령만': len(law_only),
        '저장 파일': ['final_merged_all_columns.csv', 'ordinance_only.txt', 'law_only.txt', 'common_values.txt']
    }
```
